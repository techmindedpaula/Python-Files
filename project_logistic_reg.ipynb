LOGISTIC REGRESSION

```python
data = df[['WorkLifeBalance', 'Attrition', 'RelationshipSatisfaction', 'Gender']]
```

```python
data
```

```python
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[['WorkLifeBalance', 'RelationshipSatisfaction', 'Gender']], data.Attrition, test_size = 0.2)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
