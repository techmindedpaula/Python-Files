```python
import pandas as pd
import seaborn as sns
import os
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_csv('USA_Housing.csv')
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
#data['yr_built'] = pd.to_numeric(data['yr_built'] ,errors='coerce')
```

```python
data.info()
```

```python
current_year = datetime.now().year
data['house_age'] = current_year - data['yr_built'].astype(int)
data['bathrooms_per_bedroom'] = data['bathrooms'] / data['bedrooms']
data['price_per_sqft'] = data['price'] / data['sqft_living']
```

```python
data.head()
```

```python
data[['price_per_sqft', 'house_age', 'bathrooms_per_bedroom']].head()
```

```python

cor = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'price_per_sqft', 'house_age']
correl = data[cor].corr()
```

```python
correl
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correl, annot=True, cmap='RdBu', fmt=".2f")
plt.title("Correlation")
plt.show()
```

```python
Top_10 = data.sort_values(by='price', ascending=False).head(100)
bed_age = data[data['bedrooms'] > 3].sort_values(by='house_age', ascending=True)
bed_bath = data[(data['bedrooms'] > 3) & (data['bathrooms'] >= 2)]

age_5 = data[data['house_age'] > 5]

print("Top 100 Most Expensive Properties:\n", Top_10.head())
print("Filtered for houses with more than 3 bedrooms and sorted by house age:\n", bed_age)
print("Properties with more than 3 bedrooms and 2 bathrooms:\n", bed_bath.head())
print("Properties with house age less than 5 years:\n", age_5.head())
```

```python

sorted_filtered_3 = data[data['price'] > 600000].sort_values(by='yr_built', ascending=False)
print("Filtered for houses with price > 600,000 and sorted by year built:\n", sorted_filtered_3)
```

```python

fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# 1. Distribution of Price
sns.histplot(data['price'], bins=30, ax=axes[0, 0])
axes[0, 0].set_title('Price Distribution')

# 2. Price vs. Sqft_living
sns.scatterplot(data=data, y='sqft_living', x='price', ax=axes[0, 1])
axes[0, 1].set_title('Price vs. Square Footage')

# 3. Price vs. Number of Bedrooms
sns.boxplot(data=data, x='bedrooms', y='price', ax=axes[1, 0])
axes[1, 0].set_title('Price by Number of Bedrooms')

# 4. House Age Distribution
sns.histplot(data['house_age'], bins=30, ax=axes[1, 1])
axes[1, 1].set_title('House Age Distribution')

# 5. Year Built vs Price
sns.scatterplot(data=data, x='yr_built', y='price', ax=axes[2, 0])

axes[2, 0].set_title('Year Built vs Price')

# 6. House_Age vs. Price
sns.lineplot(data=data, x='house_age', y='price', ax=axes[2, 1], ci=None)
axes[2, 1].set_title('Price Trend over House Age')

plt.tight_layout()
plt.show()
```
