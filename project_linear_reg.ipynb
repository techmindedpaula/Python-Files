```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
reg = linear_model.LinearRegression()
```

```python
df = pd.read_csv("fishers_market.csv")
```

```python
df
```

```python
reg.fit(df[['Length1', 'Length2', 'Length3', 'Height', 'Width']], df.Weight)
```

```python
predicted_weight = reg.predict(df[['Length1', 'Length2', 'Length3', 'Height', 'Width']])
```

```python
df['predicted_weight'] = reg.predict(df[['Length1', 'Length2', 'Length3', 'Height', 'Width']])

column_name = 'Weight'

cols = list(df.columns)

new_col_position = cols.index(column_name) + 1

cols.insert(new_col_position, cols.pop(cols.index('predicted_weight')))

df = df[cols]
```
